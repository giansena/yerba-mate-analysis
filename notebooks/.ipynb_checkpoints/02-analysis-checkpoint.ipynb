```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/yerba-mate-1970-2024.csv", encoding="latin-1")
```

```python
df.head()
```

```python
produccion_por_anio = df.groupby("anio")["produccion_tm"].sum()

produccion_por_anio
```

```python
produccion_por_anio.plot()
plt.title("Total yerba mate production per year")
plt.xlabel("Year")
plt.ylabel("Production (tons)")
plt.show()
```

```python
produccion_por_provincia = df.groupby("provincia")["produccion_tm"].sum()

produccion_por_provincia
```

```python
produccion_por_provincia.plot(kind="bar")
plt.title("Total production per province")
plt.show()
```

```python
top_departamentos = df.groupby("departamento")["produccion_tm"].sum().sort_values(ascending=False)

top_departamentos.head(10)
```

```python
rendimiento_por_anio = df.groupby("anio")["rendimiento_kgxha"].mean()

rendimiento_por_anio.plot()
plt.title("Average yield per year")
plt.show()
```

```python
plt.scatter(df["superficie_cosechada_ha"], df["produccion_tm"])
plt.title("Area vs. Production")
plt.xlabel("Harvested area (ha)")
plt.ylabel("Production (tonnes)")
plt.show()
```
