```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/yerba-mate-1970-2024.csv", encoding="latin-1")
```

```python
produccion_anual = df.groupby("anio")["produccion_tm"].sum()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    x=produccion_anual.index,
    y=produccion_anual.values
)

plt.title("Total yerba mate production in Argentina (1970–2024)")
plt.xlabel("Year")
plt.ylabel("Production (tonnes)")
plt.show()
```

```python
produccion_provincia = df.groupby("provincia")["produccion_tm"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=produccion_provincia.index,
    y=produccion_provincia.values
)

plt.title("Total production per province")
plt.xlabel("Province")
plt.ylabel("Total production")
plt.show()
```

```python
rendimiento_anual = df.groupby("anio")["rendimiento_kgxha"].mean()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    x=rendimiento_anual.index,
    y=rendimiento_anual.values
)

plt.title("Average yield per year")
plt.xlabel("Year")
plt.ylabel("kg per hectare")
plt.show()
```

```python
top_departamentos = (
    df.groupby("departamento")["produccion_tm"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=top_departamentos.values,
    y=top_departamentos.index
)

plt.title("Top 10 producing departments")
plt.xlabel("Total production")
plt.ylabel("Departament")
plt.show()
```
