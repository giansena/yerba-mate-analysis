```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("../data/yerba-mate-1970-2024.csv", encoding="latin-1")

df.head()
```

```python
conexion = sqlite3.connect("../data/yerba_mate.db")
```

```python
df.to_sql("produccion", conexion, if_exists="replace", index=False)
```

```python
query = "SELECT * FROM produccion LIMIT 5"

resultado = pd.read_sql(query, conexion)

resultado
```

```python
query = """
SELECT provincia, SUM(produccion_tm) as total_produccion
FROM produccion
GROUP BY provincia
"""

resultado = pd.read_sql(query, conexion)

resultado
```

```python
query = """
SELECT departamento, SUM(produccion_tm) as total_produccion
FROM produccion
GROUP BY departamento
ORDER BY total_produccion DESC
LIMIT 10
"""

resultado = pd.read_sql(query, conexion)

resultado
```

```python
query = """
SELECT anio, SUM(produccion_tm) as produccion_total
FROM produccion
GROUP BY anio
ORDER BY anio
"""

resultado = pd.read_sql(query, conexion)

resultado
```

```python
conexion.close()
```
